As we discussed in the blog, we can integrate custom css directly in one of the code cells and then, can be applied to HTML elements as below. Addtionally, the below code demonstrates how we can use media queries and make html elements responsive across various devcies.

For more details; please feel free to visit - <a href="https://sites.google.com/view/gautamsatyndra/blog/customized-and-responsive-jupyter-notebook" target="_blank"> Responsive Jupyter-HTML Page</a>. This will provide you complete steps and help you create interactive and appealing notebooks. 
<br><br>
With below contents and codes, try resizing your screens, you will notice changes on elements provided on screen sizes. 
<br><br><br><br>

```python
%%HTML

<style>
    .example-parent__div {
        font-size: 22px; 
        font-family: 'Lexend', sans-serif;
        display: flex;
        max-width: 100%;
        justify-content: center; /* you can see here that I used it as center for other example */
        align-items: center;
        padding: 50px 25px;
        justify-content: space-between;
    }

    .hello1 {
        padding: 50px;
        background-color: rgba(231, 75, 96, 1);
        border: 1px dashed #c7c6c6;
        background: linear-gradient(to bottom right, rgb(29, 12, 56) 30%, rgba(27, 15, 73, 0.25)), #EE4A60 80%;
        background-blend-mode: normal; 
        color: white;
    }

/* add media queries for smaller device */
    @media only screen and (max-width: 767px) {
        .example-parent__div {
            justify-content: center;
            flex-direction: column;
        }
    }
</style>
```
